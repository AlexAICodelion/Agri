在农产品市场的现有研究中，供需关系与价格波动机制被广泛认为是理解市场运行和滞销等现象的核心逻辑。许多实证研究表明，农产品价格波动具有明显的供需驱动特征，并且价格在不同市场类型、区域和时间尺度上呈现出明确的波动规律和动态传导机制。例如，黄先明（2011）通过对我国大宗农产品价格波动的影响因素进行了详细梳理，指出供给和需求因素是价格波动的核心驱动力，特别是在市场供需失衡的情况下，价格的波动会更加显著。张海峰强调，农产品价格波动是由供需关系变化引起的，尤其是当供应超过需求时，价格会呈现下降趋势，这为我们进一步构建供给强度和需求压力指标提供了理论依据 。
此外，王朝辉，陈洁光, 欧进锋（2021）的研究从农产品流通体系的视角分析了价格波动的机制，指出市场价格受供需关系和流通环节的双重影响，并通过分析农产品流通体系中的不同环节，揭示了市场价格波动背后的供需动态。特别是，在流通环节不完善的情况下，价格的波动幅度会加大，这与我们研究的供给强度和需求压力的构建理念非常一致。通过这两篇文献，可以进一步说明供需关系在农产品价格形成中的主导作用，从而为供给强度和需求压力这两个指标提供了强有力的支撑 。
特别在蔬菜这一典型农产品的研究中，陈巧等人（2020）分析了蔬菜批发价格与零售价格之间的传导机制，指出批发价格在价格形成过程中的主导作用，并强调了供需变化对价格波动的影响。该研究表明，批发价格的波动往往反映了市场供需关系的变化，尤其是在供需失衡时，价格波动会加剧，进而影响消费者和农民收入。此外，Yang, Y., Li, B., & Wang, M. (2020) 的研究表明，需求信号对于价格波动有着至关重要的作用，尤其是当市场需求压力较大时，价格会上涨并影响市场的供需关系。
根据这些文献，本文构建了供给强度和需求压力两个指标，并通过它们来量化市场滞销现象的严重程度。供给强度（批发上市量 / 批发价格）反映了供给与价格之间的关系，而需求压力（批发价格 / 产地价格）则反映了市场需求对价格的影响。通过这两个指标，我们能够量化市场供需失衡并为进一步研究滞销与收入下降之间的关系提供量化基础。
